```python


data_fake['date'] = pd.to_datetime(data_fake['date'], format='%B %d, %Y', errors='coerce')
data_fake['date'] = data_fake['date'].combine_first(pd.to_datetime(data_fake['date'], format='%d-%b-%y', errors='coerce'))
data_fake['month'] = data_fake['date'].dt.month
data_fake['day'] = data_fake['date'].dt.day
data_fake['year'] = data_fake['date'].dt.year

data_fake
```

```python
data_fake.dropna(subset=['month'], inplace=True)
data_fake.dropna(subset = ['day'], inplace=True)
data_fake.dropna(subset = ['year'], inplace=True)

data_fake
```

```python
data_fake['month'] = data_fake['month'].astype(int)
data_fake ['day'] = data_fake['day'].astype(int)
data_fake['year'] = data_fake['year'].astype(int)

data_fake
```

```python
data_true
```

```python
data_true['date'] = pd.to_datetime(data_true['date'])
data_true['month'] = data_true['date'].dt.month
data_true['day'] = data_true['date'].dt.day
data_true['year'] = data_true['date'].dt.year

data_true
```

```python
news = pd.concat([data_fake, data_true], ignore_index=True)
news
```

```python
news_count_by_date = news.groupby('date').size().reset_index(name='news_count')
news_count_by_date
```
